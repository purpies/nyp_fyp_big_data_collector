```python
import urllib.request
import urllib
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import os
import time
```

```python
#url = "https://ezbuy.sg/category/562?sortCondition=5&ezspm=1.20000006.3.0.562"
#url = "https://shopee.sg/search?keyword=men%20slipper&page=1"
#driver.get(url)

driver = webdriver.Chrome("C:\\Users\\L31206\\Desktop\\Shopee Slipper Webscrape\\chromedriver.exe")
driver.maximize_window()
#soup = BeautifulSoup(driver.page_source, "html.parser")

productDataSaved = ""
pageNo = 0

for pageNo in range(60, 100):
    
    soup = driver.get("https://shopee.sg/search?keyword=men%20slipper&page=" + str(pageNo) )
    time.sleep(2)
    myElem = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.CLASS_NAME, 'shopee-search-item-result__items')))
    
    elm = driver.find_element_by_tag_name('html')
    elm.send_keys(Keys.END)
    time.sleep(1)
    elm = driver.find_element_by_tag_name('html')
    elm.send_keys(Keys.END)
    time.sleep(1)
    elm = driver.find_element_by_tag_name('html')
    elm.send_keys(Keys.END)
    time.sleep(1)
    
    soup = BeautifulSoup(driver.page_source, "html.parser")
    try:
        #soup = BeautifulSoup(driver.page_source, "html.parser")
        
        for product in soup.findAll('div',{"class":"shopee-search-item-result__item"}):
            time.sleep(1)
            productData = ""
            
            detailedItemLink = product.find('a')
            try:
                #Mining unique HREF link
                driver.get("https://shopee.sg" + detailedItemLink.attrs['href'] )
                hrefLink = str(detailedItemLink.attrs['href'])
            except AttributeError:
                print("hrefError")

            try:
                time.sleep(1.5)
                elm = driver.find_element_by_tag_name('html')
                elm.send_keys(Keys.END)
                time.sleep(1.5)
                myElem = WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CLASS_NAME, 'BtHdNz')))
                myElem = WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CLASS_NAME, 'qaNIZv')))
                myElem = WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CLASS_NAME, '_3Lybjn')))
                soup = BeautifulSoup(driver.page_source, "html.parser")
                
                #Product Name
                productName = soup.find('div',{"class":"qaNIZv"}).text.replace(',','')
                
                #Current Selling Price/Discounted Price
                SellingPrice = soup.find('div',{"class":"_3n5NQx"}).text.replace(',','')
                
                #Before Discounted Price (if any)
                try:
                    BeforeDiscountedPrice = soup.find('div',{"class":"_3_ISdg"}).text.replace(',','')
                    DiscountPercentage = soup.find('div',{"class":"MITExd"}).text.replace(',','')
                    
                except AttributeError:
                    BeforeDiscountedPrice = "N.A."
                    DiscountPercentage = "N.A."
                    
                #Bundle deals (if any)
                try:
                    bundleOffer = soup.find('div',{"class":"_3MJQdO"}).text.replace(',','')
                except AttributeError:
                    bundleOffer = "No Bundles Avaliable"
                
                #Free Shipping?
                try:
                    freeShipping = soup.find('div',{"class":"_5y6zF4"}).text.replace(',','')
                    
                except AttributeError:
                    freeShipping = "N.A."
                
                #Shipping Fee
                
                ShippingFee = soup.find('div',{"class":"BtHdNz"}).text.replace(',','')
                
                #Quantity Sold
                QuantitySold = soup.find('div',{"class":"_22sp0A"}).text.replace(',','')
                
                #Brand
                try:
                    Brand = soup.find('a',{"class":"_2H-513"}).text.replace(',','')
                except AttributeError:
                    Brand = "No brand"
                
                #Ratings with photo
                try:
                    photoRatingCount = soup.find('div',{"class":"product-rating-overview__filter--with-photo"}).text.replace(',','')
                except AttributeError:
                    photoRatingCount = "N.A"
                    
                #ratings with comment
                try:
                    commentRatingCount = soup.find('div',{"class":"product-rating-overview__filter--with-comment"}).text.replace(',','')
                except AttributeError:
                    commentRatingCount = "N.A"
            
                
                #Overall Product rating over 5
                try:
                    ProductRating = soup.find('span',{"class":"product-rating-overview__rating-score"}).text.replace(',','')
                except AttributeError:
                    ProductRating = "No Ratings Yet"
                
                #Stock Count (how many quantity left?)
                StockCount = soup.find('div',{"class":"_1FzU2Y"}).text.replace(',','')
                
                #Store Info (6 Details)
                StoreInfoArray = []
                for StoreInfo in soup.findAll('span', {"class": "_1rsHot"}):
                    StoreInfoArray.append(StoreInfo.text.replace(',',''))
                
                #Popularity (How many added to favourites/wishlist)
                wishlistCount = soup.find('div',{"class":"_25DJo1"}).text.replace(',','')
                    
                #Seller/Store Info
                storeName = soup.find('div',{"class":"_3Lybjn"}).text.replace(',','')
                
                productData = productData + "," + hrefLink + "," + productName + "," + SellingPrice + "," + BeforeDiscountedPrice + "," + DiscountPercentage + "," + bundleOffer + "," + freeShipping + "," + ShippingFee + "," + QuantitySold + "," + Brand + "," + ProductRating + "," + photoRatingCount + "," + commentRatingCount + "," +wishlistCount + "," + storeName + "," + ";".join(StoreInfoArray)
            
            except TimeoutException:
                print("Individual Item Loading took too much time!")
                
            if len(productData) != 0:
                productDataSaved = productDataSaved + "\n" + productData[1:]

            header = "Product Link,Product Name,Selling Price,Price Before Discount,DiscountPercentage,Bundle Offer,Free Shipping,Shipping Price,Quantity Sold,Brand,Overall Rating,Rating with Photo,Rating with comments,Amount in Wishlist,Store Name,Store Details"
            file = open(os.path.expanduser("ShoppeeSlipperDataDemo.csv"), "wb")
            file.write(bytes(header, encoding="ascii", errors='ignore'))
            file.write(bytes(productDataSaved, encoding="ascii", errors='ignore'))
            
            
        
    except TimeoutException:
        print("Loading took too much time!")
        
    pageNo += 1 
        #driver.back();
               
        
    
```
