```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
srxWebsite = 'https://www.srx.com.sg/search/sale/hdb/'

#Use PD to read every stret name in SG. Replace all whitespaces in street to "+". Redirect link is srxWebsite + street with "+".
streetDataset = r'C:\Users\L31206\Desktop\Address Scrape (SRX)\Street.xlsx'
df_streetDataset = pd.read_excel(streetDataset)
streetNameList = df_streetDataset.Street.tolist()

addressDataSaved = ""

for eachStreet in streetNameList:
    
    editedStreet = eachStreet.replace(' ' , '+')
    
    pagelink = "https://www.srx.com.sg/search/sale/hdb/"+editedStreet
    page = requests.get(pagelink)
    propertyPage = BeautifulSoup(page.content, 'html.parser')
    
    for individualProperty in propertyPage.findAll('div',{"class":"listingDetailsDiv"}):
        
        individualPropertyLink = individualProperty.find('a')
        
        detailedIndividualProperty = requests.get("https://www.srx.com.sg" + individualPropertyLink.attrs['href'] )
        detailedIndividualPropertySoup = BeautifulSoup(detailedIndividualProperty.content, 'html.parser')
        
        addressSegmentList = detailedIndividualPropertySoup.findAll('div',{"class":"listing-about-main-value"})
        
        try:
            address = addressSegmentList[6].text
        except IndexError:
            address = " "
        
        print(address)
        
        addressData = eachStreet + "," + address
        
        if len(addressData) != 0:
            addressDataSaved = addressDataSaved + "\n" + addressData[0:]

        header = "Street Name, Address"
        file = open(os.path.expanduser("street_plus_address.csv"), "wb")
        file.write(bytes(header, encoding="ascii", errors='ignore'))
        file.write(bytes(addressDataSaved, encoding="ascii", errors='ignore'))
            
print("Address Scrape Completed")
```
