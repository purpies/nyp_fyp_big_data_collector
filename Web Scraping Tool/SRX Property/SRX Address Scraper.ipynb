```python
import requests
from bs4 import BeautifulSoup

srxWebsite = 'https://www.srx.com.sg/hdb'
```

```python
page = requests.get(srxWebsite)

soup = BeautifulSoup(page.content, 'html.parser')

for address in soup.findAll('li',{"class":"hdb-search-item"}):
    linkHref = address.find('a')
    
    drilldownLink = requests.get("https://www.srx.com.sg" + linkHref.attrs['href'] )
    soupDrilldown = BeautifulSoup(drilldownLink.content, 'html.parser')
    
    soupDrilldown_UL = soupDrilldown.find('ul',{"id":"hdb-town-streets-list"})
    
    for detailedAddress in soupDrilldown_UL.findAll('li'):
        print(detailedAddress.a.text)
```
